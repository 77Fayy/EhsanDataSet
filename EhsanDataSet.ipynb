```python

import arabic_reshaper
import bidi

def arabic_text(text):
    reshaped_text = arabic_reshaper.reshape(text)
    bidi_text = bidi.algorithm.get_display(reshaped_text)
    return bidi_text

year_counts = combined['العام'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(year_counts, labels = year_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20.colors)

plt.title(arabic_text("جمع التبرعات السنوية لمنصة إحسان"))
plt.show()
```

```python

plt.figure(figsize=(10, 6))

donations['InitiativeName_display'] = donations['InitiativeName'].apply(lambda x: get_display(arabic_reshaper.reshape(x)))

sns.lineplot(data=donations, x='InitiativeName_display', y='DonationCount', marker='o')

plt.title(get_display(arabic_reshaper.reshape("تبرعات في منصة إحسان")))
plt.xlabel(get_display(arabic_reshaper.reshape("نوع التبرع")))
plt.ylabel(get_display(arabic_reshaper.reshape("تكرار التبرع")))
plt.xticks(rotation=90)

plt.show()
```

```python
filtered_data = donations[donations['PaymentMethod'].isin(['أبل باى', 'بطاقة الدفع'])]
donations['PaymentMethod'] = donations['PaymentMethod'].str.strip()
```

```python

plt.figure(figsize=(10, 6))
ax = sns.countplot(x='PaymentMethod', data=filtered_data, palette='Spectral')

plt.title(arabic_text("توزيع طرق الدفع: أبل باى مقابل بطاقة الدفع"))
plt.xlabel(arabic_text("طريقة الدفع"))
plt.ylabel(arabic_text("عدد التبرعات"))

xticks_labels = filtered_data['PaymentMethod'].unique()

xticks_labels_arabic = [arabic_text(label) for label in xticks_labels]


ax.set_xticklabels(xticks_labels_arabic, rotation=45, horizontalalignment='right')

plt.show()
```
